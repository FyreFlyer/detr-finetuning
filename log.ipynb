```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the log.txt file
with open('log.txt', 'r') as file:
    lines = file.readlines()

# Parse each line as a dictionary and store it in dataframe
data = []
for line in lines:
    entry = eval(line)
    data.append(entry)

df = pd.DataFrame(data)

# print(df)
# print(df.columns)

# Plotting the train_loss column
plt.plot(df.index, df['train_loss'], label='Train Loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend()

plt.show()


# Plot test_class_error
plt.plot(df['epoch'], df['test_class_error'])

plt.xlabel('Epoch')
plt.ylabel('Test Class Error')
plt.title('Test Class Error vs Epoch')

plt.show()
```
